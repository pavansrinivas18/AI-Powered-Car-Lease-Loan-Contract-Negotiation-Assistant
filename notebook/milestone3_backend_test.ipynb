```python
# Simulate the Backened Contract Lookup (Upload API Logic)

def get_contract_by_id(id):
    record =df[df["id"]==id]

    if record.empty:
        return None
    return record.iloc[0].to_dict()
```

```python
# Negotiation Logic

def negotiation_advice(contract):
    advice = []

    if contract["interest_rate"] > 10:
        advice.append("Try negotiating interest rate")

    if contract["tenure_months"] > 60:
        advice.append("Ask for shorter tenure")

    if contract["risk_flag"] == "high":
        advice.append("Proceed carefully or compare other offers")

    if not advice:
        advice.append("Contract terms look reasonable")

    return advice
```

```python
contract = get_contract_by_id(5)
negotiation_advice(contract)
```

```python
final_output = {
    "contract_id" : contract["id"],
    "customer_name" : contract["customer_name"],
    "vehicle_type" : contract["vehicle_type"],
    "monthly_emi" : contract["monthly_emi"],
    "interest_rate" : contract["interest_rate"],
    "tenure_months" : contract["tenure_months"],
    "risk_flag" : contract["risk_flag"],
    "negotiation_advice" : negotiation_advice   
}

final_output
```
