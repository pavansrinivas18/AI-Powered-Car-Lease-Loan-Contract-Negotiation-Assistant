```python
import pandas as pd

contracts = pd.read_csv("../data/sample_car_contracts_with_vin.csv")
prices = pd.read_csv("../data/market_prices.csv")

contracts.head()
```

```python
import sys
import os

# Add project root directory to Python path
PROJECT_ROOT = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(PROJECT_ROOT)

print("Project root added:", PROJECT_ROOT)
```

```python
from src.price_engine import estimate_fair_price
from src.fairness import contract_fairness_score
```

```python
final_results = []

for _, row in contracts.iterrows():
    fair_price = estimate_fair_price(
        row["vehicle_type"],
        row["monthly_emi"],
    )

    fairness = contract_fairness_score(
        row["clause_summary"],
        fair_price,
        row["monthly_emi"]
    )

    final_results.append({
        "customer_name": row["customer_name"],
        "vehicle_type": row["vehicle_type"],
        "monthly_emi": row["monthly_emi"],
        "fair_price": fair_price,
        "fairness_score": fairness,
        "risk_flag": row["risk_flag"],
        "issue_type": row["issue_type"],
        "recommended_action": row["recommended_action"],
        "vin": row["vin"]
    })

final_df = pd.DataFrame(final_results)
final_df.head()
```
